```python
students = 5
sum = 0

for i in range(students):
    score = int(input(f'請輸入第{i+1}位學生的成績:'))
    sum += score

print(f"全班總成績為:{sum},平均為{sum/students}") #字串插補
print("全班總成績為:%d,平均為:%.2f" % (sum, sum/students)) #格式化字串(舊的寫法)
print("全班總成績為:{:.1f},平均為:{:.2f}".format(sum, sum/students)) #格式化字串(新的寫法)
```

```python
for i in range(1,10):
    for j in range(1,10):
        print(f"{i}*{j}={i*j}", end=' ')
    print("\n")
```

```python
for i in range(1,10):
    for j in range(1,10):
        print("{:d} * {:d}={:2d}".format(i,j,i*j), end=' ')
    print("\n")
```

```python
for i in range(1,10):
    if i>5 :
        break;
    for j in range(1,10):
        print(f"{i}*{j}={i*j}", end=' ')
    print("\n")
```

```python
for i in range(1,10):
    if i % 2 == 0 :
        continue
    for j in range(1,10):
        print(f"{i}*{j}={i*j}", end=' ')
    print("\n")
```
