```python
import os
import csv

root = os.path.abspath('./')
filePath = os.path.join(root,"csv_data","個股日成交資訊.csv")

with open(filePath,encoding='utf-8',newline='') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    stock_data = list(csv_reader)

print(stock_data)
print(len(stock_data))
    
```
