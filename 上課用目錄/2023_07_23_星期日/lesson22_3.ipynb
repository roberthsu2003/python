```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips
```

```python
tips.info()
```

```python
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
grouped = tips.groupby(by=['日期','吸煙者'])
len(grouped)
```
