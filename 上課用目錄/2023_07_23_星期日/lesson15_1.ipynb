```python
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac: font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac: font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window: font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

student1 =  {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}

figure = plt.figure(figsize=(8, 5), dpi=250, facecolor="#cccccc")
axes = figure.add_subplot(1,1,1)
axes.plot(student1.keys(),student1.values(),'g--o',label='stu1')
axes.plot(student2.keys(),student2.values(),'r-+',label='stu2')
axes.set_xticklabels(student1.keys(),fontproperties=font)
axes.legend()
axes.set_title("學生1和學生2分數比較表",fontproperties=font)
axes.set_xlabel("科目",fontproperties=font)
axes.set_ylabel("分數",fontproperties=font)
#axes.annotate('90',(1,70))
for x in axes.get_xticks():
    value1 = list(student1.values())[x]
    value2 = list(student2.values())[x]
    print(x,value1)
    axes.annotate(str(value1),(x+0.05,value1),fontproperties=font,color='red',
                  arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', 
                            color='b'))
    axes.annotate(str(value2),(x+0.05,value2),fontproperties=font,color='green',
                  arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', 
                            color='b'))
```

```python
import random
import matplotlib.pyplot as plt
#days = [num for num in range(1,31)]
days = list(range(1,31))
celsius = [round(random.uniform(24.3, 37.5),ndigits=1) for _ in range(1,31)]
figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(days,celsius,'r--.',label='September')
axes.grid()
axes.legend()
axes.set_title("Weather")
axes.set_xlabel("Days")
axes.set_ylabel("Celsius")
axes.set_xticks(days)
plt.show()
```
