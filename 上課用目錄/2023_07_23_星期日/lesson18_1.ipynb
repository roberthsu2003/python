```python
import pandas as pd
import numpy as np

df = pd.read_csv('world.csv')
df
```

```python
df.info()
```

```python
allDataFrame = df.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
taiwan = allDataFrame[allDataFrame['國家'] == '台灣']
taiwan
```

```python
taiwan1 = allDataFrame[(allDataFrame['國家'] == '台灣') & (allDataFrame['新增確診數'] == 0)]
taiwan1
```

```python
taiwan2 = allDataFrame.query('國家=="台灣"')
taiwan2
```

```python
taiwan3 = allDataFrame.query('國家=="台灣" and 新增確診數==0')
taiwan3
```

```python
max = allDataFrame.query('國家=="台灣"')["新增確診數"].max()
allDataFrame.query('國家=="台灣" and 新增確診數==@max')
```

```python
max = allDataFrame.query('國家=="日本"')["新增確診數"].max()
allDataFrame.query('國家=="日本" and 新增確診數==@max')
```

```python
max = allDataFrame.query('國家=="美國"')["新增確診數"].max()
allDataFrame.query('國家=="美國" and 新增確診數==@max')
```
