```python
file = open('names.txt',mode='r',encoding='utf-8')
names:list[str] = []
for line in file:
    #names.append(line)
    line = line.rstrip('\n')
    names.append(line)
file.close()

len(names)
```

```python
#亂數抓一個姓名
import random
first_name = "徐"
for _ in range(5):
    random_name = random.choice(names)
    print(first_name + random_name[1:])
```
