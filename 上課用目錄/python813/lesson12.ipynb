```python
#figure
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,5),facecolor="white")
axes = figure.add_subplot(1,1,1)
x = [0, 1, 10]
y = [3, 4, 5]
axes.plot(x,y)
plt.show()
```

```python
figure = plt.figure(figsize=(8,5),dpi=220,facecolor="#cccccc")
axes1 = figure.add_subplot(1,3,1)
axes2 = figure.add_subplot(1,3,2)
axes3 = figure.add_subplot(1,3,3)

axes1.set_title("chart 1")
axes2.set_title("chart 2")
axes3.set_title("chart 3")

plt.show()
```

```python
import matplotlib.pyplot as plt

student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':72, 'math':87, 'history':61, 'discover':90}

subject = list(student1.keys())
student1_score = list(student1.values())
student2_score = list(student2.values())

figure = plt.figure(figsize=(8,5),dpi=220,facecolor='#cccccc')
axes = figure.add_subplot()
axes.plot(subject,student1_score)
axes.plot(subject,student2_score)

plt.show()
```
