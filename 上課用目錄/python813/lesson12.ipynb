```python
import matplotlib.pyplot as plt

student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':72, 'math':87, 'history':61, 'discover':90}

subject = list(student1.keys())
student1_score = list(student1.values())
student2_score = list(student2.values())

figure = plt.figure(figsize=(8,5),dpi=220,facecolor='#cccccc')
axes = figure.add_subplot()
axes.plot(subject,student1_score,'m--D')
axes.plot(subject,student2_score,'r-.o')

plt.show()
```

```python
#
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

subject = ['','','','','']
figure = plt.figure(figsize=(8,5),dpi=220,facecolor='#cccccc')
axes = figure.add_subplot()
axes.plot(subject,student1_score,'m--D')
axes.plot(subject,student2_score,'r-.o')
axes.set_xticklabels(subject,fontproperties=font)
axes.set_title("12",fontproperties=font)
axes.legend(['1','2'],prop=font)
for i in range(len(subject)):
    axes.annotate(str(student1_score[i]),(i,student1_score[i]+0.8),fontproperties=font)
    axes.annotate(str(student2_score[i]),(i,student2_score[i]+0.8),fontproperties=font)

axes.annotate(":\n",(3,85),fontproperties=font)
plt.show()
```
