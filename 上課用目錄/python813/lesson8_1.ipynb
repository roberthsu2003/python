```python
#台北市youbike2.0及時資料
#https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json
import requests
youbike_taipei = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
```

```python
response = requests.get(youbike_taipei)
print(type(response))
```

```python
if response.status_code == 200:
    print("下載成功")
```

```python
#json文字檔
json_txt = response.text
```

```python
#將文字變為file物件
import io
file = io.StringIO(json_txt)
```

```python
#透過json module建立python的資料結構
import json
data = json.loads(json_txt)
data
```
