# 1. 單向選擇
只在條件為 True 時執行

```python
score = 85

if score >= 60:
    print("及格！")
```

# 2. 雙向選擇
條件為 True 時執行第一段，否則執行第二段

```python
age = 15

if age >= 18:
    print("已成年")
else:
    print("未成年")
```

# 3. 多項選擇
依序判斷多個條件

```python
score = 75

if score >= 90:
    print("優等")
elif score >= 80:
    print("甲等")
elif score >= 70:
    print("乙等")
elif score >= 60:
    print("丙等")
else:
    print("丁等")
```
