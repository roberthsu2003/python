# Practice 8: 猜數字遊戲

這個練習將製作一個簡單的猜數字遊戲，練習以下概念：
- 迴圈 (while)
- 條件判斷 (if/elif/else)
- 使用者輸入 (input)
- 隨機數 (random)

```python
import random

# 遊戲設定
secret_number = random.randint(1, 100)  # 產生 1~100 的隨機數
max_attempts = 10  # 最多猜測次數
attempts = 0  # 目前猜測次數

print("歡迎來到猜數字遊戲！")
print(f"我想了一個 1 到 100 之間的數字，你有 {max_attempts} 次機會猜中它！")
print("-" * 40)
```

```python
# 遊戲主迴圈
low = 1  # 目前範圍下限
high = 100  # 目前範圍上限

print("\n遊戲開始！我會提示你答案的範圍喔～")
print("-" * 40)

while attempts < max_attempts:
    try:
        # 顯示目前範圍
        remaining = max_attempts - attempts
        print(f"\n目前範圍：{low} ~ {high} (剩余 {remaining} 次機會)")
        print(f"數線：{'.' * (low-1)}|{'=' * (high-low+1)}|{'.' * (100-high)}")
        print(f"        {low}{' ' * (high-low-1)}{high}")
        
        guess = int(input("請輸入你的猜測："))
        
        if guess < low or guess > high:
            print(f"請輸入 {low} 到 {high} 之間的數字！")
            continue
        
        attempts += 1
        
        if guess == secret_number:
            print(f"\n🎉 恭喜你！猜對了！答案就是 {secret_number}")
            print(f"你總共猜了 {attempts} 次")
            break
        elif guess < secret_number:
            print(f"⬆️  太小了！答案在 {guess+1} 到 {high} 之間")
            low = guess + 1  # 縮小範圍下限
        else:
            print(f"⬇️  太大了！答案在 {low} 到 {guess-1} 之間")
            high = guess - 1  # 縮小範圍上限
            
    except ValueError:
        print("請輸入有效的數字！")
else:
    print(f"\n😢 遊戲結束！你已經用完 {max_attempts} 次機會")
    print(f"正確答案是：{secret_number}")
```
