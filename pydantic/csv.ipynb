```python
#轉換個股日成交資訊
from pydantic import RootModel,BaseModel,Field
import csv
from pprint import pprint

class Stack(BaseModel):
    代號:str = Field(alias='證券代號')
    名稱:str = Field(alias='證券名稱')
    成交股數:str
    成交金額:str
    開盤價:str
    最高價:str
    最低價:str
    收盤價:str
    漲跌價差:str
    成交筆數:str


class Stacks(RootModel):
    root: list[Stack]

    def __iter__(self):
        return iter(self.root)
    
    def __getitem__(self,item):
        return self.root[item]


with open('個股日成交資訊.csv',encoding='utf8', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    #for row in reader:
    #    print(row)
    all = Stacks(reader)

for item in all:
    pprint(item)
```
