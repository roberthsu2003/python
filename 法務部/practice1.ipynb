### `print()` 練習挑戰
請嘗試在下方的單元格中實作以下功能：

| 功能描述 | 提示關鍵字 |
| :--- | :--- |
| 同時印出 "年齡:" 與數字 25 | 使用逗號 `,` |
| 印出 "A", "B", "C" 並用 "|" 隔開 | 使用 `sep="|"` |
| 連續呼叫兩次 print，但結果顯示在同一行 | 使用 `end=""` |

```python
# 在這裡練習你的 print() 範例
print("年齡:" , 25)
print("A", "B", "C",sep="-")
print("年齡:",end="")
print(25)
```
