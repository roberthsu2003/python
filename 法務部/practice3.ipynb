```python
# 要求學生分數最高為 300 分，符合條件加分 5%，如果超過 300 則以 300 分計

score = int(input("請輸入學生分數(最高300分): "))
if score <= 300:
  is_add = input("學生是否符合加分條件?(y/n): ")

  if is_add == "y":
    score *= 1.05

    if score > 300:
      score = 300

else:
  print("學生分數不可以大於 300 分！")

print(f"學生分數是:{round(score)}")
print("應用程式結束")
```
