```python
#有參數沒有傳出值
def func_sum(a, b):
    c = a + b
    print(c)

func_sum(3, 5)
```

```python
#有參數又有傳出值的
def func_sum(a,b):
    c = a + b
    return c

sum = func_sum(5, 8)
print(sum)
```

```python
def temperature(value):
    return 1.8 * value + 32

while(True):
    value = int(input('請輸入攝氏溫度:'))
    result = temperature(value)
    print(f'華氏溫度{result}')
    
    inputValue = input('程式還要繼續嗎?(輸入N...結束):')
    if inputValue == 'N':
        break;
    
print("程式結束")        
```
