```python
#輸入一個值,如果是偶數,不作任何動作,奇數輸出平方值,按'q'離開程式
while True:
    value = input("請輸入整數,輸入'q'離開:")
    if value == 'q':
        break
    number = int(value)
    if number % 2 == 0:
        continue
    
    print(f"{number}的平方是:{number ** 2}")
print("程式結束")
```

```python
#巢狀迴圈
for i in range(1,10):
    for j in range(1,10):
        print(f"{i:2d}*{j:2d}={i*j:2d}",end=" ")
    print()
```
