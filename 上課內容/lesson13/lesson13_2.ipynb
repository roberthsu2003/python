```python
import csv

file = open('各鄉鎮市區人口密度.csv',encoding='utf-8')
reader = csv.reader(file)
dicData = dict()
for row in reader:    
    if row[0] == '106':
        try:
            dicData[row[1]] = int(row[2])
        except:
            dicData[row[1]] = 0
print(dicData)        
file.close()
```

```python
import pandas as pd
obj = pd.Series(dicData)
print(obj)
```

```python
for row in obj[obj<=10000].index:
    print(row)
```
