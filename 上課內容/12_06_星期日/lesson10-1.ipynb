```python
#讀取csv檔
import csv

with open('students2.csv', 'r', encoding='utf-8') as file:
    csvReader = csv.reader(file)
    print(type(csvReader))
    studentsList = list(csvReader)
    for item in studentsList:
        print(item)
        if item[0] == 'A00012':
             for score in item[1:]:
                    print(score)
```

```python
import csv
with open('各鄉鎮市區人口密度.csv', 'r', encoding='utf8') as file:
    csvReader = csv.reader(file)
    poplist = []
    for one in csvReader:
        if one[0] == '106':
            poplist.append(one)
    
    for one in poplist:
        if '永和區' in one[1]:
            print(one)
```

```python
#儲存為json文件

import json
codes = {'tw':'Taiwan', 'jp':'japan', 'hk':'HongKong','us':'美國'}
jsonStr=json.dumps(codes, ensure_ascii=False)
print(jsonStr)

with open('codes.json','w',encoding='utf-8') as file:
    file.write(jsonStr)
print('json檔案存檔完成')
```

```python
import random
import json
students = []
for i in range(1,51):
    stdDict = {'姓名':'A1000' + str(i),
              '國文':random.randint(50,100),
              '英文':random.randint(50,100),
              '數學':random.randint(50,100),
              '自然':random.randint(50,100),
              '社會':random.randint(50,100),}
    students.append(stdDict)
with open('students.json', 'w', encoding='utf8') as file:
    json.dump(students,file, ensure_ascii=False)
print('json存檔成功')
```

```python
#請取json檔案
import json
with open('students.json', 'r', encoding='utf-8') as file:
    students = json.load(file)
for student in students:
    print(student)
```
