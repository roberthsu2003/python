```python
#傳統函數 

def f(x, y, z):
    return x+y+z;

print(f(10,20,30))
```

```python
#匿名函數
noNameFunction = lambda x,y,z:x+y+z
print(noNameFunction(10,20,30))
```

```python
#傳統函數 

def square(x):
    return x ** 2

newList = list(map(square,[1,2,3,4,5]))
newList
```

```python
#匿名函數
newList = list(map(lambda x:x ** 2,[1,2,3,4,5]))
newList
```
